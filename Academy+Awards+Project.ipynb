```python
#reading the data into a pandas dataframe
import pandas as pd
academy_awards = pd.read_csv('C:\\Users\\Samuel\\python3stuff\\academy_awards.csv', encoding ='ISO-8859-1',dtype={"Year" : "object"})
academy_awards.head(20)
```

```python
#seeing the type of each column
academy_awards.info()



```

```python
# only keeping the year, not the number of the ceremony
academy_awards["Year"] = academy_awards["Year"].str[0:4]
academy_awards.head()
```

```python
#changing year column to integers and filtering to keep only years after 2000
academy_awards['Year'] = academy_awards['Year'].astype('int64')
later_than_2000 = academy_awards[academy_awards['Year']>2000]
```

```python
#filtering out only the result for the categories we are interested in and after 2000
award_categories = ['Actor -- Leading Role', 'Actor -- Supporting Role', 'Actress -- Leading Role',
'Actress -- Supporting Role']
nominations = later_than_2000[later_than_2000["Category"].isin(award_categories)]
```

```python
#replacing yes and no in the Won? column so it is easier to query later
pd.options.mode.chained_assignment = None # default='warn'
no_yes = { "NO": 0, "YES": 1 }
nominations["Won?"] = nominations["Won?"].map(no_yes)
```

```python
#adding a column called Won
nominations['Won'] = nominations['Won?']
```

drop_columns = ['Won?','Unnamed: 5','Unnamed: 6','Unnamed: 7','Unnamed: 8','Unnamed: 9','Unnamed: 10']
final_nominations = nominations.drop(drop_columns, axis = 1)

```python
#dropping erroneous columns
drop_columns = ['Won?','Unnamed: 5','Unnamed: 6','Unnamed: 7','Unnamed: 8','Unnamed: 9','Unnamed: 10'] 
final_nominations = nominations.drop(drop_columns, axis = 1)
final_nominations
```

```python
#removes brackets from additional info column, and seperates the movie and characters into different columns
additional_info_one = final_nominations["Additional Info"].str.rstrip("'}")
additional_info_two = additional_info_one.str.split(" {'")
movie_names = additional_info_two.str[0]
characters = additional_info_two.str[1]
final_nominations["Movie"] = movie_names
final_nominations["Character"] = characters
final_nominations = final_nominations.drop("Additional Info", axis=1)
final_nominations.head()
```

```python
#turning our pandas dataframe into an sql database
import sqlite3
conn = sqlite3.connect('nominations.db')
final_nominations.to_sql("nominations", conn, index=False)
```

```python
#quering the database and printing some results
query_one = "pragma table_info(nominations);"
query_two = "SELECT * from nominations LIMIT 10"
schema = conn.execute(query_one).fetchall()
first_ten = conn.execute(query_two).fetchall() 

print(schema)
for r in first_ten:
    print(r)

```

```python
#creating a new table called ceremonies with hosts and years of different academy awards
years_hosts = [(2010, "Steve Martin"),
               (2009, "Hugh Jackman"),
               (2008, "Jon Stewart"),
               (2007, "Ellen DeGeneres"),
               (2006, "Jon Stewart"),
               (2005, "Chris Rock"),
               (2004, "Billy Crystal"),
               (2003, "Steve Martin"),
               (2002, "Whoopi Goldberg"),
               (2001, "Steve Martin"),
               (2000, "Billy Crystal"),
            ]
create_ceremonies = "CREATE TABLE ceremonies (id integer primary key, year integer, host text);"
conn.execute(create_ceremonies)
insert_query = "insert into ceremonies (Year, Host) values (?,?);"
conn.executemany(insert_query, years_hosts)
```

```python
#checking to make sure our new table works
print(conn.execute('SELECT * from ceremonies LIMIT 10').fetchall())
print(conn.execute("pragma table_info(ceremonies);").fetchall())
```

```python
#turning froeign key constraints
conn.execute("PRAGMA foreign_keys = ON;")
```

```python
#creating a new table that is th same as nominations but without the year column and a ceremony_id column
create_nominations_two = '''CREATE TABLE nominations_two (id integer primary key, 
category text, 
nominee text, 
movie text, 
character text, 
won integer,
ceremony_id integer,
foreign key(ceremony_id) references ceremonies(id));
'''
nom_query = '''
select ceremonies.id as ceremony_id, nominations.category as category, 
nominations.nominee as nominee, nominations.movie as movie, 
nominations.character as character, nominations.won as won
from nominations
inner join ceremonies 
on nominations.year == ceremonies.year
;
'''
joined_nominations = conn.execute(nom_query).fetchall()

conn.execute(create_nominations_two)

insert_nominations_two = '''insert into nominations_two (ceremony_id, category, nominee, movie, character, won) 
values (?,?,?,?,?,?);
'''

conn.executemany(insert_nominations_two, joined_nominations)
print(conn.execute("select * from nominations_two limit 5;").fetchall())
```

```python
#deleting nominations tabe and renaming nomintions_two to nominations
delete_nominations = 'DROP TABLE nominations;'
conn.execute(delete_nominations)
rename = 'ALTER TABLE nominations_two RENAME TO nominations;'
conn.execute(rename)
```

```python
#creating new empty tables
movies_query = 'CREATE TABLE movies (id integer PRIMARY KEY, movie text)'
actor_query = 'CREATE TABLE actors (id integer PRIMARY KEY, actor text)'
movie_actor_query ='''CREATE TABLE movies_actors (id integer PRIMARY KEY, movie_id integer references movies(id), 
actor_id integer references actors(id));
'''
conn.execute(movies_query)
conn.execute(actor_query)
conn.execute(movie_actor_query)
```

```python
#inserting data into the movie and actors table
insert_movies = "insert into movies (movie) select distinct movie from nominations;"
insert_actors = "insert into actors (actor) select distinct nominee from nominations;"
conn.execute(insert_movies)
conn.execute(insert_actors)

print(conn.execute("select * from movies limit 5;").fetchall())
print(conn.execute("select * from actors limit 5;").fetchall())
```

```python
#inserting data into the join table
pairs_query = "select movie,nominee from nominations;"
movie_actor_pairs = conn.execute(pairs_query).fetchall()

join_table_insert = "insert into movies_actors (movie_id, actor_id) values ((select id from movies where movie == ?),(select id from actors where actor == ?));"
conn.executemany(join_table_insert,movie_actor_pairs)

print(conn.execute("select * from movies_actors limit 5;").fetchall())
```
